Compare a random agent with an agent trained on optimal policies with 5 blocks setting

```python
from modelEvaluator import modelEvaluator
from experiment import experiment
from experiment import experimentSet
```

```python
eS = experimentSet()
numBlocks = 5
numEpisodes = 30
```

```python
from RandomAgent import RandomAgent

exp = experiment('Random Agent')

randomAgent = RandomAgent (numBlocks)
mE = modelEvaluator(randomAgent)

trial = mE.runEpisodesForAgent(numEpisodes,numBlocks)
exp.addTrial(trial)
eS.addExperiment(exp)

#eS.plotExperimentSet()
```

```python
from keras_seq2seq_agent import BasicSeq2SeqAgent

basicseq2seqAgent = BasicSeq2SeqAgent(numBlocks)
mE = modelEvaluator(basicseq2seqAgent)

trial = mE.runEpisodesForAgent(numEpisodes,numBlocks)
exp = experiment('BasicSeq2SeqAgent')
exp.addTrial(trial)
eS.addExperiment(exp)
```
